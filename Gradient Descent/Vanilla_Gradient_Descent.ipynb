# Defining The Function

## Convex Function

### Function 1

```python
def f1(x,first_order=False,second_order=False):
    m=(x[0]**2+2*x[1]**2)
    if first_order==True:
        m1=np.array([2*x[0],4*x[1]])
        m1.reshape(2,1)
        return m,m1
    if second_order==True:
        m2.array([2,0,0,-4])
        m2.reshape(2,2)
        return m,m2
    return m
```

## RosenBrock Function

```python
def ros(x,first_order=False,second_order=False):
    m=(x[0]**2 - x[1])**2  + (1 - x[0])**2
    if first_order==True:
        m1=np.array([4*x[0][0]*(x[0][0]**2 - x[1][0]) - 2*(1-x[0][0]), (-2)*(x[0][0]**2 - x[1][0])]).reshape(2,1)
        return m,m1
    if second_order==True:
        m2=np.array([[12*x[0][0]**2-4*x[1][0]+2,-4*x[0][0]],[-4*x[0][0],2]]).reshape(2,2)
        return m,m2
    else:
        return m
```

## Vanilla Gradient Descent

```python
def vanilla_gradient_descent(start_point,learning_rate,error,max_count):
    _,direction=f1(start_point,first_order=True)
    direction=direction.reshape(2,1)
    point2 = start_point-learning_rate*(direction)
    l=[start_point,point2]
    count=1
    while np.linalg.norm(l[-1]-l[-2])>error:
        count=count+1
        if count>max_count:
            break
        else:
            _,direction=f1(l[-1],first_order=True)
            point=l[-1]-learning_rate*(direction)
            l.append(point)
    return l,count
```

```python
l2,c2=vanilla_gradient_descent(x,0.1,0.00001,10000)
```
